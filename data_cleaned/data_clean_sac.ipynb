 calculate percent increase from yr to yr in sac
*  create df and define func (function)
*  create copy 
   *  copy() method creates a copy of an existing list

```python
def calculate_percent_increase(x1, x2):
    return ((x2-x1)/x1)*100
```

```python
sac_percent = sac.copy()
for i in range(sac_percent.shape[1]-1,1, -1):
    prior = sac_percent.iloc[:, i-1]
    current = sac_percent.iloc[:, i] 
    sac_percent.iloc[:, i] = calculate_percent_increase(prior, current)
```

 Let's convert the index to datetime!

*  strftime() function is used to convert to Index using specified date_format.
*   columns.name=none- names the index, by default it is None or blank, the columns are of type IndexArray

```python
sac_percent.index = p.DatetimeIndex(sac_percent.index)
sac_percent.index = sac_percent.index.strftime('%m/%Y')
sac_percent.columns.name = None
```

```python
sac_price.index = p.DatetimeIndex(sac_price.index)
sac_price.index = sac_price.index.strftime('%m/%Y')
sac_price.columns.name = None
```
