#### calculate percent increase from yr to yr in sac

```python
def calculate_percent_increase(x1, x2):
    return ((x2-x1)/x1)*100
```

```python
sac_p = sac.copy()
for i in range(sac_p.shape[1]-1,1, -1):
    prior = sac_p.iloc[:, i-1]
    current = sac_p.iloc[:, i] 
    sac_p.iloc[:, i] = calculate_percent_increase(prior, current)
```

#### look at percent increase

```python
sac_p.head()
```

#### transpose data

```python
sac_p = sac_p.set_index('RegionName').transpose()
```

```python
sac = sac.set_index('RegionName').transpose()
```

#### fix datetime

```python
sac_p.index = p.DatetimeIndex(sac_p.index)
sac_p.index = sac_p.index.strftime('%m/%Y')
sac_p.columns.name = None
```

```python
sac.index = p.DatetimeIndex(sac.index)
sac.index = sac.index.strftime('%m/%Y')
sac.columns.name = None
```

#### exporting ü•í


```python
with open('pickles/timeseries_sac_p.pkl', 'wb') as fp:
    pickle.dump(sac_p, fp)
```

```python
with open('pickles/timeseries_sac.pkl', 'wb') as fp:
    pickle.dump(sac, fp)
```
