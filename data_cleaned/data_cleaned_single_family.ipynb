 calculate percent increase from yr to yr in sac
*  create df and define func (function)
*  create copy 
   *  copy() method creates a copy of an existing list

```python
def calculate_percent_increase(x1, x2):
    return ((x2-x1)/x1)*100
```

```python
single_family_percent = single_family.copy()
for i in range(single_family_percent.shape[1]-1,1, -1):
    prior = single_family_percent.iloc[:, i-1]
    current = single_family_percent.iloc[:, i] 
    single_family_percent.iloc[:, i] = calculate_percent_increase(prior, current)
```

```python
single_family_percent.head(3)
```

```python
single_family_percent.drop('2000-01-31' ,axis=1, inplace=True)
```

 Let's convert the index to datetime!

*  strftime() function is used to convert to Index using specified date_format.
*   columns.name=none- names the index, by default it is None or blank, the columns are of type IndexArray

```python
single_family_percent.index = p.DatetimeIndex(single_family_percent.index)
single_family_percent.index = single_family_percent.index.strftime('%m/%Y')
single_family_percent.columns.name = None
```

```python
single_family_price.index = p.DatetimeIndex(single_family_price.index)
single_family_price.index = single_family_price.index.strftime('%m/%Y')
single_family_price.columns.name = None
```
